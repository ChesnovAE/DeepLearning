```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer

from keras import Model, Sequential
from keras.layers import InputLayer, Dropout, Dense, BatchNormalization, Activation
from keras.optimizers import Adam

import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
df = pd.read_csv('Churn_Modelling.csv', index_col='RowNumber')
```

```python
df.head()
```

```python
obj_cols = []
```

```python
for i, dtype in enumerate(df.dtypes):
    if dtype == 'object':
        obj_cols.append(df.columns[i])
```

```python
obj_cols
```

```python
class AnnClassifier:
    def __init__(self, input_dim):
        self.input_dim = input_dim
        self.model = Sequential()
        self.model.add(InputLayer(input_shape=(self.input_dim,)))
        self.model.add(Dense(8, activation='relu'))
        self.model.add(Dense(4, activation='relu'))
        self.model.add(Dense(2, activation='relu'))
        self.model.add(Dense(1, activation='sigmoid'))

        self.model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    
    def fit(self,
            X_train,
            y_train,
            batch_size,
            epochs):
        self.model.fit(X_train, y_train, batch_size=batch_size, epochs=epochs)
    
    def predict(self, X):
        return self.model.predict(X)
```

```python
_df = df.drop('Surname', axis=1)
```

```python
x_df = df.drop(['Surname', 'Exited', 'CustomerId'], axis=1).copy()
y_df = df['Exited'].copy()

x_train, x_test, y_train, y_test = train_test_split(x_df, y_df, test_size=0.33, random_state=142)
```

```python
x_train.shape
```

### Data preprocess

```python
for col in ['Gender', 'Geography']:
    le = LabelEncoder()
    x_train[col] = le.fit_transform(x_train[col])
    x_test[col] = le.transform(x_test[col])

ct = ColumnTransformer([('one_hot_encoder', OneHotEncoder(categories='auto'), [1])], remainder='passthrough')
x_train = ct.fit_transform(x_train)
x_test = ct.transform(x_test)

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
ann = AnnClassifier(input_dim=x_train.shape[1])
```

```python
batch_size = 128
epochs = 100
ann.fit(x_train, y_train, batch_size, epochs)
```

```python
import lightgbm as lgb
```

```python
train_ds = lgb.Dataset(x_train, y_train)
params = {
    'boosting_type': 'gbdt',
    'objective':'binary',
    'num_leaves': 31,
    'learning_rate': 0.05,
    'feature_fraction': 0.9,
    'bagging_fraction': 0.8,
    'bagging_freq': 5,
    'verbose': 0
}
gbm = lgb.train(params, train_ds, num_boost_round=100)
gbm_preds = gbm.predict(x_test)
ann_preds = ann.predict(x_test)
```

```python
print(classification_report(y_test, (gbm_preds > 0.5)))
```

```python
print(classification_report(y_test, (ann_preds > 0.5)))
```
